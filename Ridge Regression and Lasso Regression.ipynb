```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import sklearn 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge, LinearRegression, Lasso
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
```

```python
pd.set_option('display.max_columns',None)
```

```python
sales = pd.read_csv('home_data.csv',index_col = 0)
```

```python
sales.info()
```

```python
sales.describe()
```

### <div class="alert alert-block alert-success"> Part I  - Polynomial Regression</div>

```python
X=sales[['sqft_living']]
```

```python
y = sales.price
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state = 0)
```

```python
pipe = make_pipeline(PolynomialFeatures(degree=15),StandardScaler(),LinearRegression())
pipe.fit(X_train, y_train)
```

```python
print(f"best trainval-set score with poly features: {pipe.score(X_train, y_train):.3f}")
print(f"best test-set score with poly features: {pipe.score(X_test, y_test):.3f}")
```

### <div class="alert alert-block alert-success"> Part II  - Ridge Regression</div>

```python
pipe = make_pipeline(PolynomialFeatures(degree=15),StandardScaler(),Ridge())

param_grid = {'ridge__alpha': np.logspace(3, 9, 13)}

grid = GridSearchCV(pipe, param_grid, cv=10, return_train_score=True)
grid.fit(X_train, y_train)
```

```python
results = pd.DataFrame(grid.cv_results_)
results
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
grid.score(X_test, y_test)
```

### <div class="alert alert-block alert-success"> Part III  - Lasso Regression</div>

#### <div class="alert alert-block alert-success"> Part 1</div>

```python
from math import log, sqrt
```

```python
sales['sqft_living_sqrt'] = sales['sqft_living'].apply(sqrt)
sales['sqft_lot_sqrt'] = sales['sqft_lot'].apply(sqrt)
sales['bedrooms_square'] = sales['bedrooms']*sales['bedrooms']
sales['floors_square'] = sales['floors']*sales['floors']
```

```python
X = sales[['bedrooms', 'bedrooms_square', 'bathrooms', 'sqft_living', 'sqft_living_sqrt', \
     'sqft_lot', 'sqft_lot_sqrt', 'floors', 'floors_square', 'waterfront', 'view', \
     'condition', 'grade', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated']]
```

```python
y=sales.price
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.1,random_state=0)
```

```python
pipe2 = make_pipeline(StandardScaler(),Lasso(max_iter=1e6,alpha=100))

lasso1=pipe2.fit(X_train,y_train)
lasso1
```

```python
sum(lasso1['lasso'].coef_ != 0)
```

```python
X_train.columns[lasso1["lasso"].coef_ != 0] #choosen features
```

```python
X_train.columns[lasso1["lasso"].coef_ == 0] #left out feature 
```

```python
print(f'best test-set score: {lasso1.score(X_test, y_test):.3f}')
```

#### <div class="alert alert-block alert-success"> Part 2</div>

```python
pipe3 = make_pipeline(StandardScaler(),Lasso(max_iter=1e6))

param_grid = {'lasso__alpha': np.logspace(1, 7, 13)}

grid3 = GridSearchCV(pipe3, param_grid, cv=10, return_train_score=True, n_jobs= -1)
grid3.fit(X_train, y_train)
```

```python
results = pd.DataFrame(grid3.cv_results_)
results.head()
```

```python
print(f'best Î»:{grid3.best_params_}')
print(f'best CV score:{grid3.best_score_:.3f}')
lasso2 = grid3.best_estimator_
lasso2
```

```python
print(X_train.shape)
np.sum(lasso2['lasso'].coef_ != 0)
```

```python
lasso_coef_2 = lasso2['lasso'].coef_
lasso_coef_2
```
